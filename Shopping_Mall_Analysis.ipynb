```python
dp = pd.read_csv(r"C:\Users\DELL\Desktop\Shopping_Mall_Cleaned.csv")
dp
```

```python
Average_annual_income = dp['Annual Income'].mean()
Average_annual_income
```

```python
Average_spending_score = dp['Spending Score'].mean()
Average_spending_score
```

```python
def get_segment(row):
    if row['Annual Income'] > Average_annual_income and row['Spending Score'] > Average_spending_score:
        return "High Income High Spending"

    elif row['Annual Income'] > Average_annual_income and row['Spending Score'] <= Average_spending_score:
        return "High Income Low Spending"

    elif row['Annual Income'] <= Average_annual_income and row['Spending Score'] > Average_spending_score:
        return "Low Income High Spending"

    else:
        return "Low Income Low Spending"


dp['segment'] = dp.apply(get_segment, axis=1)
```
